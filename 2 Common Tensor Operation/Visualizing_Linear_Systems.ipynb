## **Visualizing Linear Systems**

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000) # start, finish, n points
```

```python
y1 = 2 * x + 1
```

```python
y2 = 3 * x - 5
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([0, 5])
ax.set_ylim([0, 8])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='red')
plt.axvline(x=2, color='black', linestyle='--')
_ = plt.axhline(y=6, color='black', linestyle='--')
```
